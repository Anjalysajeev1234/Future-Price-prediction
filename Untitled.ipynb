```python
#TASK:4
#PREPARE THE DATA BEFORE APPLYING FACEBOOK PROPHET TOOL

avocado_df
avocado_prophet_df=avocado_df[['Date','AveragePrice']]
        
```

```python
avocado_prophet_df=avocado_prophet_df.rename(columns={'Date':'ds','AveragePrice':'y'})
```

```python
avocado_prophet_df
```

```python
#understand the intution behind facebook prophet
# prophet is an opensource software released by Facebook's core Data Science Team
# Prophet is a aprocedure for forecasting time series data based on an additive model where non linear trends are fit with yearly,weekly,and daily seasonality plus holiday effects

#TASK:6
#DEVELOP MODEL AND MAKE PREDICTIONS

m=Prophet()
m.fit(avocado_prophet_df)

#forecasting into the future

future=m.make_future_dataframe(periods=365)
forecast=m.predict(future)
```

```python
forecast
```

```python
figure=m.plot(forecast,xlabel='Date',ylabel='Price')
```

```python
figure_2=m.plot_components(forecast)
```
